**File Loading**

```python
with open("B:/PsychoStasis/Training/DexterChatML.txt", "r") as file:
  fileLines = file.readlines()  
  
print(len(fileLines))
```

**Remove s .. /s**

```python
fileLines = list(map(lambda line: line.replace("<s>","").replace("</s>", "").strip(), fileLines))
fileLines[0:10]
```

**Join small lines with the previous ones**
and
***Remove empty lines***

```python
for i in range(len(fileLines)):
  line = fileLines[i].strip()
  if(len(line)<5):
    fileLines[i] = ""
  elif ((len(line)< 20) or (str(line[0]) == str(line[0]).lower())): 
    fileLines[i-1] = fileLines[i-1] + " " + line
    fileLines[i] = ""
  
    
fileLines = list(filter(lambda line: len(line) > 0, fileLines))

print(len(fileLines))

  
```

**Std Deviation and Quartiles computation**

```python
import pandas as pd

lines = list(map(lambda line: line.strip(), fileLines))

frame = pd.DataFrame(lines, columns=["text"])
frame["wordCount"] = frame["text"].apply(lambda x: len(x.split()))
  
frame.head()

stdDev = frame['wordCount'].std()
print(f"Std. Dev.: {stdDev}")

Q1 = frame['wordCount'].quantile(0.25)
print('First Quartile (25th Percentile):', Q1)

Q3 = frame['wordCount'].quantile(0.75)
print('Third Quartile (75th Percentile):', Q3)
```

***Quartiles Verification***

```python
overQ3 = frame[frame['wordCount'] > Q3]
overQ3Count = len(overQ3)
print(f"Over 75th Percentile: {overQ3Count}")

underQ1 = frame[frame['wordCount'] < Q1]
underQ1Count = len(underQ1)
print(f"Under 25th Percentile: {underQ1Count}")
fileLines[1:10]
```

***Split long paragraphs***

```python
procLines = []
for line in fileLines:
  wordCount = len(line.split(" "))
  if(wordCount< Q3):
    procLines.append(line)
  else:
    upperSide = ""
    lowerSide = ""
    lineList = line.split('.')
    lineList = list(map(lambda ln: ln.strip()+".", lineList))
    listLen = len(lineList)
    upperSide = str.join(' ', lineList[:int(listLen/2)])
    lowerSide = str.join(' ', lineList[int(listLen/2):])
    procLines.append(upperSide)
    procLines.append(lowerSide)
    
fileLines = procLines
print(len(fileLines))

fileLines
```

**Junction and removal of small lines**

```python
for i in range(len(fileLines)):
  line = fileLines[i].strip()
  if(len(line)<5):
    fileLines[i] = ""
  elif ((len(line)< 20) or (str(line[0]) == str(line[0]).lower())): 
    fileLines[i-1] = fileLines[i-1] + " " + line
    fileLines[i] = ""
  
    
fileLines = list(filter(lambda line: len(line) > 20, fileLines))

print(len(fileLines))
```

**Line formatting**

```python
for i in range(len(fileLines)):
  fileLines[i] = f"<s>{fileLines[i].strip()}</s>\n"

fileLines[0:10]
```

**File Saving**

```python
with open("B:/PsychoStasis/Training/Dexter-ext.txt", "w") as file:
  for line in fileLines:
    file.write(line.strip()+"\n")
```
